## Helper

```python
def inspect(label, data):
    """Print top-level keys, shape (if tabular), and first 2 records."""
    print(f"\n{'='*60}")
    print(f"  {label}")
    print(f"{'='*60}")

    if isinstance(data, dict):
        print(f"Top-level keys : {list(data.keys())}")
        # find the first list value to treat as rows
        rows = next((v for v in data.values() if isinstance(v, list)), None)
        if rows is None:
            rows = [data]          # whole dict is one record
    elif isinstance(data, list):
        print(f"Top-level type : list")
        rows = data
    else:
        print(f"Value : {data}")
        return

    df = pd.json_normalize(rows)
    print(f"Shape          : {df.shape}")
    print(f"\nFirst 2 records:")
    print(df.head(2).to_string(index=False))
```

## 1 — Historical Round-Level Data
`historical-raw-data/rounds` · tour=pga · event_id=14 · year=2024

```python
rounds = pull_endpoint(
    "historical-raw-data/rounds",
    {"tour": "pga", "event_id": "14", "year": "2024"}
)
inspect("historical-raw-data/rounds  (pga | event 14 | 2024)", rounds)
```
